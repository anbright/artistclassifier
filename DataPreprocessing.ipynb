```python
path = 'data/all_data_info.csv'
df = pd.read_csv(path, header=0)
```

```python
artists = ['Claude Monet',
           'Pablo Picasso',
           'Paul Cezanne',
           'Claude Monet',
           'Camille Pissarro',
           'Rembrandt',
           'Henri Matisse']
```

```python
artist_count = df.groupby('artist').count()
top50 = artist_count.sort_values(by='genre', ascending=False).head(25)
top50['genre'].index.tolist()
top50.to_csv('data/top_artists.csv', index=True)
```

```python
artist_paintings = pd.DataFrame(columns=['artist', 'title', 'in_train','filename'])
artist_df = df.loc[df['artist'].isin(artists)]
# artist_df[['artist', 'title', 'in_train', 'new_filename']].to_csv('data/artist_files.csv')
artist_df
```

```python
import os
# Saves the subset of the data related to the above artists
training = artist_df.loc[artist_df['in_train'] == True][['artist','new_filename']]

for index, row in training.iterrows():
    name = row['artist'].lower().replace(' ', '_')
    filename = row['new_filename']
    newfilename = name + '_' + filename
    try:
        os.rename('train/'+filename, 'data/train/'+newfilename)
    except Exception as e:
        print(e)
    
```
